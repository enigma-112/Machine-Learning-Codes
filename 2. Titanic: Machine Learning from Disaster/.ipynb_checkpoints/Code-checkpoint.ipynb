```python
def entropy(column):
    entropy_ans=0
    N= float(column.shape[0])  # N = total no. of rows (total no. of data points)
    items=np.unique(column,return_counts=True)
        
    for i in items[1]:
        p = i/N
        entropy_ans += (-1.0)*p*np.log2(p) # p = probability of each class
    
    return entropy_ans
```
